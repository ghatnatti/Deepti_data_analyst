```python
###Replace missing by mean and mode for numeric and object respectively



for column in df:
    if df[column].isnull().any():
        if(column in numeric_columns):
            df[column]=df[column].fillna(df[column].mean())
            
for column in df:
    if df[column].isnull().any():
        if(column in category_columns):
            df[column]=df[column].fillna(df[column].mode()[0])
        
df.info()
```

```python
#####Independent sample T test between all numeric independent variables and a categorical dependent variable Churn. 
###To test if there is significant mean difference between the variables
```

```python
i=0
while True:
    print("Mean Difference:",df[numeric_columns[i]].groupby(df['Churn']).mean())
    t,p=stats.ttest_ind(df[numeric_columns[i]][df['Churn']==0],df[numeric_columns[i]][df['Churn']==1])
    print("t=",t,  "p=",p)
    if(p<=0.05):
        print(numeric_columns[i], "and Churn have statistically significant mean difference")
    else:
        print(numeric_columns[i], "and Churn don't have statistically significant mean difference")
    print("####################################")
    i=i+1
    if (i >= len(numeric_columns)):
        break
```

```python
#####Chi sq test between all catagorical independent variables and a categorical dependent variable Churn.
###To test the association between variables
```

```python
category_columns=df.select_dtypes(include=['object']).columns.tolist()
```

```python
i=0
while True:
    data=pd.crosstab(df[category_columns[i]],df['Churn'],normalize='columns')
    print(data)
    c, p, dof, expected = chi2_contingency(data)
    print("p value of",category_columns[i],"is :",p )
    if(p<0.05):
        print(category_columns[i], "and Churn are significantly associated")
    else:
        print(category_columns[i], "and Churn are not significantly associated")
    print("########################################")        
    i=i+1
    if (i>=len(category_columns)):
        break
```
